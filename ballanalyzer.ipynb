```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import onedquad as quad
```

```python
mod = quad.createball()
mod.setparam('ground.e',10.0)
mod.setparam('ground.fr',0.4)
mod.setparam('frame.v',10)
```

```python
mod.getallparams()
```

```python
sym = pd.DataFrame(mod.simulate(0.0,40.0,0.0001))
```

```python
sym
```

```python
plt.plot(sym[['t']],sym[['framex']])
```
